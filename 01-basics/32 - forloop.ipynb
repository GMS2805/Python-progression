```python
print("A CONTAINER can be a range of numbers, a string of characters, or a list of values.")
print("A for loop is capable of iterating over objects within a container.")

# Example - For loop within a string
str_var = "A string"

count = 0
for c in str_var:
    count += 1

print(count)

print()
print("The range function is a commonly used tool for counting inside a for loop")
print("In the range function, the start, end and step size can be modified")
#Examples of how to use the range function
# range(end) = range(4) gives 0,1,2,3
# range(start,end) = range(2,6) gives 2,3,4,5
# range(start,end,step) = range(1,7,2) gives 1,3,5

# Exercise - Write a program using a for loop that takes in a string as input and counts the number of spaces in the provided string.
sentence = input("Enter a sentence:")

count = 0
for c in sentence:
    if c == ' ':
        count += 1
        
print(count)

```
