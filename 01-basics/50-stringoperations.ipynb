```python
print("A string is a sequence of characters")
print("String characters can be compared using logical operators ><!= and membership operators in and not in.")
print("To compare two characters, the matching characters are compared sequentially until a decision is reached.")
print("To compare characters by logical operators, their ASCII values are compared.")
print("The info below shows how the operators work.")

"""
> or >=
Checks whether the first string value is greater than (or greater than or equal to) the second string value.	
"c" > "d" gives False
When comparing "c" operand to "d" operand, the ASCII value for "c" is smaller than the ASCII value for "d". Therefore, "c" < "d". The expression "c" > "d" evaluates to False.


< or <=
Checks whether the first string value is less than (or less than or equal to) the second string value.	
"ab" < "ac" gives True
When comparing "ab" operand to "ac" operand, the first characters are the same, but the second character of "ab" is less than the second character in "ac" and as such "ab" < "ac".


==
Checks whether two string values are equal.	
"aa" == "aa" outputs True
Since all characters in the first operand and the second operand are the same, the two string values are equal.


!=
Checks whether two string values are not equal.	
"a" != "b" outputs True
The two operands contain different string values ("a" vs. "b"), and the result of checking whether the two are not the same evaluates to True.

in
Checks whether the second operand contains the first operand.	
"a" in "bc" would output False
Since string "bc" does not contain string "a", the output of "a" in "bc" evaluates to False.

not in
Checks whether the second operand does not contain the first operand.	
"a" not in "bc" outputs True
Since string "bc" does not contain string "a", the output of "a" not in "bc" evaluates to True.
"""

print("aaa"<"aab")
print("aa"<"a")
print("ples" in "apples")
print()
print("upper and lower functions can convert strings into lowercase or uppercase letters. They do NOT modify the string.")

#Example
x = "Apples"

# The lower() method converts a string to all lowercase characters
print(x.lower())

# The upper() method converts a string to all uppercase characters
print(x.upper())

print("aBbA".lower())
print("aBbA".upper())
print("a".upper() == "A")

# Exercise - Create a variable s_input, input string and output the upper and lowercase version of the string
s_input =  input("What's on your mind?")
print(s_input.lower())
print(s_input.upper())
print(len(s_input))

# Exercise - if the character is between "a" and "t" or "A" and "T", print True. Otherwise, print False.
if s_input.lower() >= 'aa' and s_input.lower() <= 'tz':
    print("True")
else:
    print("False")

print("What is clear here, is that the condition is evaluating only the first character of the if condition entered.")
print("If i use two characters in condition, it evaluates two characters. First character gets priority.")
print("This can extend to as many characters as i like in the string.")
```
