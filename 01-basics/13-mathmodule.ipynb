```python
print("A module is a previously written code which can be imported into a program.")
print("Python has an extensive standard library of modules")
print("The import statement defines a variable for accessing code from a specific module")
print("Import statements are at the start of the program")
print("Functions such as print, input, float are built-in functions. These are always available and importing these is not needed.")
print("The complete list of built-in functions is listed at https://docs.python.org/3/library/functions.html ")
print("The math module is a commonly used module and includes functions such as sqrt, which are common math operators.")
print()
print("Example program to calculate distance between two points")
import math
    
x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))
    
distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)
print("The distance is", distance)

print()
print("In a python shell, functions like sqrt and constants like pi can be accessed by math followed by a dot.")
print("Commonly used math functions are detailed below;")
print("math.e is euler's number, the base of the natural logarithm. e = 2.71828..")
print("math.pi is the ratio of circumference to diameter of a circle. pi = 3.141..")
print("math.tau is the circumference - radius ratio. It is two times pi. tau = 6.282..")
print("math.ceil(x) is the ceiling of x. the smallest integer greater than x. if x is 1.2, ceil of x would be 2.")
print("math.floor(x) is the floor of x. the biggest integer lesser than x. if x is 1.2, floor of x would be 1.")
print("math.sin(x),cos(x),tan(x). Trigonometric functions. Note - x is in radians.")
print("math.log(x) is the log of x base e. natural logarithm.")
print("math.log(x,base) is the log of x to the given base.")
print("math.pow(x,y) is x raised to the power y.")
print("math.sqrt(x) is the square root of x.")
print()

print("Exercise : Code for the quadratic formula. Python does not have a plus-or-minus, so the solutions must be output as two separate ones. One plus and the other minus.")
a = float(input("in ax^2 + bx + c = 0, enter a"))
b = float(input("Enter b"))
c = float(input("Enter c"))
x1 = (-b + math.sqrt((math.pow(b,2))-(4*a*c)))/(2*a)
x2 = (-b - math.sqrt((math.pow(b,2))-(4*a*c)))/(2*a)
print("the solutions are :",x1,"and",x2)

print()
print("Exercise: Code to calculate the volume and surface area of a cylinder, with radius and height as inputs")
r = float(input("enter radius in meters"))
h = float(input("enter height of cylinder in meters"))
v = math.pi*(math.pow(r,2))*h
sa = 2*math.pi*r*h
print("The volume of cylinder in m^2 is :",v)
print("The surface area of the cylinder in m^2 is:",sa)
```
