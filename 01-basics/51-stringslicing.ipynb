```python
print("strings are made up characters indexed from left to right, starting at 0.")
print("if the string has s characters, the left-most character is indexed 0 and the right-most character is indexed len(s) - 1.")
print("The length of the string Mukund is 6, so the last index is 5.")
print("Negative indexing can also be used to indicate chars right to left starting at -1 to -s")
print("String slicing is used to access a sequence of characters within a string.")
print("When [a:b] is used with the name of a string variable, characters from index a to index b-1 is returned.")

#Example
"""
Consider a time value is given as "hh:mm" with "hh" representing the hour and "mm" representing the minutes.
To retrieve only the string's minutes portion, the following code can be used:

    time_string = "13:46"
    minutes = time_string[3:5]
    print(minutes)
The output is 46.

Consider a time value is given as "hh:mm" with "hh" representing the hour and "mm" representing the minutes.
To retrieve only the string's hour portion, the following code can be used:

    time_string = "14:50"
    hour = time_string[:2]
    print(hour)
The above code's output is 14.
"""

print("String objects are immutable and cannot be modified or changed once created.")
print("to make changes to a string object, a new string object with the desired changes must be created.")

"""
string_variable = "example"
string_variable[-1] = ""
This outputs a type error because strings are immutable
"""
# Practice
"""Given the string "Hello my fellow classmates" containing a greeting message:
print the first word by getting the beginning of the string up to (and including) the 5th character.
Change the first word in the string to "Hi" instead of "hello" and print the greeting message again.
"""
greeting = "Hello my fellow classmates"
print(greeting[:5])
print("Hi" + greeting[5:])

#Exercise
"""
Given a string variable, string_variable, and a list of indexes:
remove characters at the specified indexes and print the resulting string.

    Input:
    string_variable = "great"
    indices = [0, 1]
    Prints: 
    eat
"""
string_variable = "great"
indices = [0, 1]

for i in indices:
    string_variable = string_variable[:i] + '*' + string_variable[i+1:]

result = ""
for i in string_variable:
    if i != '*':
        result += i

print(result)
```
