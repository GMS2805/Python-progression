```python
print("A function is a block of code which can be called at any time to carry out a task")
# print() is also a function.
# float(), input(), math.ceil(), etc are all functions.

print("A function is defined using the def keyword")
# Example def meow() :  def functionname () : is the syntax.
# later, you can call the function by using meow()

#concept - guess the output
def print_phone_num():
  print("Phone: (", 864, ")", 555, "-", 1199)

print("User info:")
print_phone_num()

#it prints user info: 864 555 0199
# output - it printed an error that leading zeros in decimal enteger literals are not permitted.
# if the number is 1199 instead of 0199, does it print? yes.

print("The benefit of a function is that it lets statements of one task be modularly set in one group in the program.")
print("It lets this task be reusable and ok to modify easily, lowers overall program length.")

""" Exercise - Write a function, concessions(), that prints the food and drink options at a cinema.

Given:

    concessions()
The output is:

    Food/Drink Options:
    Popcorn: $8-10
    Candy: $3-5
    Soft drink: $5-7
"""
def concessions():
    print("Food/Drink Options:")
    print("Popcorn: $8-10")
    print("Candy: $3-5")
    print("Soft drink: $5-7")

concessions()

""" Exercise - Write a function, terms(), that asks the user to accept the terms and conditions, reads in Y/N, and outputs a response. 
In the main program, read in the number of users and call terms() for each user.

Given inputs 1 and "Y", the output is:

    Do you accept the terms and conditions?
    Y
    Thank you for accepting.
Given inputs 2, "N", and "Y", the output is:

    Do you accept the terms and conditions?
    N
    Have a good day.
    Do you accept the terms and conditions?
    Y
    Thank you for accepting.
"""
def terms():
    print("Do you accept the terms and conditions?")
    y = input("Enter Y if yes")
    if y == "Y":
        print("Terms accepted.")
    else:
        print("Terms not accepted")

count = int(input("How many users?"))
for i in range(0, count):
    terms()
```
