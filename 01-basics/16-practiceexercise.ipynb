```python
"""Baking bread
Uses recipe by King Arthur flour
Calculates the necessities to feed guests"""

# Read user input
loaf_weight = float(input("Input size of bead loaf in grams"))
serving_size = float(input("Input approximate serving size per guest in grams"))
guests = int(input("How many guests will be present?"))

# Calculating number of loaves needed
loaves_exact = float(serving_size*guests/loaf_weight)
loaves_2d = round(loaves_exact,2)
pieces = int(4*loaves_2d)
full_loaf = pieces//4
quarter_loaf = pieces%4

# Measuring ingredients required
yeast = round(float(pieces*1.5/4),2)
salt = round(float(pieces*1.5/4),2)
sugar = round(float(pieces*1.5/4),2)
flour = round(float(pieces*2.5/4),2)
sourdough = round(float(pieces*2/4),2)
water = round(float(pieces*0.5/4),2)

# Output the required ingredients
print("To be precise,",loaves_2d,"loaves will be needed")
print("We can approximate it as",full_loaf,"full loaves and",quarter_loaf,"quarter loaves.")
print()
print("Ingredient estimates to make the bread:")
print("1 1/2 teaspoons instant yeast per loaf =",yeast,"teaspoons")
print("1 1/2 teaspoons salt per loaf =",salt,"teaspoons")
print("1 1/2 teaspoons sugar per loaf =",sugar,"teaspoons")
print("2 1/2 cups all-purpose flour per loaf =",flour,"cups")
print("2 cups sourdough starter per loaf =",sourdough,"cups")
print("1/2 cup lukewarm water per loaf =",water,"cups")
print()
print("Happy cooking!")
```
