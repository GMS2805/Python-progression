```python
print("Hello, GitHub")
print("Hello", "GitHub")
print("Hello", "Github", sep="...")
print("Hello",end=">>")
print("Github")
print("Hello","GitHub", sep=">>>>", end="**")
print("A comma inside the print brackets gives a space between the words. the sep will make each of them separated by whatever symbols are needed, and the end will let you continue printing on the same line. Without end, the next print statement goes to the next line.")
```
