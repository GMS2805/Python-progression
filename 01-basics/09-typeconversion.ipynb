```python
print("The python interpreter implicitly converts data assigned into its corresponding type. for example, 5.3 will be assigned a float type automatically.")

# example code
book_rating = 3
print("Rating:", book_rating, type(book_rating))
update = 2
book_rating = (book_rating+update)/2
print("Rating:", book_rating, type(book_rating))
print(int(book_rating))

print("As you see above, int of a float will give the lower digit. int of both 2.1 and 2.8 will become 2.")
print("int function converts things into int, float into float, and str into string. The string turns numbers or any values into just their characters.")

# exercise problem
exam_1 = float(input("Enter the first exam score: "))
exam_2 = float(input("Enter the second exam score: "))
exam_3 = float(input("Enter the third exam score: "))

average = (exam_1 + exam_2 + exam_3) / 3
print(float(average))
print(int(average))

# exercise problem 2
ounces = int(input()) 
print('Input ounces:', ounces, ', type:', type(ounces))

cups_drank = ounces / 8
cups_left = int(8 - cups_drank)
print(cups_drank, 'cups drank, type:', type(cups_drank))
print('About', cups_left, 'cups left to drink, type:', type(cups_left))
```
