```python
print("Merry Christmas!")
print()
print("A typical print statement combines words and punctuation using commas, operators and quote marks")
print("This makes it a bit difficult to read.")
print("An f-string, or a formatted string, is a way to simplify this")
print("It is prefixed with f or F, and has a replacement field.")
print("A replacement field is an expression in curly brackets.")
print()
print("Demo example:")
name = input("What is your name?")
people = input("Who do you live with?")
print(f"Merry Christmas, {name} and {people}!")
print()
print("Formatting numbers")
import math
print("You may need to display numbers up to certain digits, or with zeros in front of them")
print("For example, you might want to display time as 9:06, not 9:6.")
print("In an f-string, you can use a format specifier to adjust decimals shown or zeros shown.")
print("Example demos of different formats are below:")
print(f"{12345678:d}"" d gives default integer format")
print(f"{12345678:,d}"" ,d gives comma separators")
print(f"{12345678:10d}"" 10d gives an integer at least 10 characters wide")
print(f"{12345678:010d}"" 010d gives an integer at least 10 characters wide, with leading zeros")
print(f"{math.pi:f}"" f gives six decimal places as default")
print(f"{math.pi:.4f}"" .4f is a fixed-point rounded to 4 decimal places")
print(f"{math.pi:8.4f}"" 8.4f is a fixed-point rounded to 4 decimal places and has a width of at least 8 characters")
print(f"{math.pi:08.4f}"" 8.4f is a fixed-point rounded to 4 decimal places and has a width of at least 8 characters and leading zeros")
print()
# Exercise
print("Wage calculator using inputs as in time, out time and hourly rate")
start_timehr = int(input("Start time hour: "))
start_timemin = int(input("Start time minute: "))
stop_timehr = int(input("Stop time hour: "))
stop_timemin = int(input("Stop time minute: "))
rate = float(input("Hourly wage: "))
print()

start = start_timehr*60 + start_timemin
stop = stop_timehr*60 + stop_timemin
hours = (stop - start) / 60
payment = hours * rate

print(f"Working hours were from {start_timehr}:{start_timemin:02d} to {stop_timehr}:{stop_timemin:02d}")
print(f"Time clocked: {hours:.1f}")
print(f"Wage earned: ${payment:.2f}")
```
